```python
#EDA-1-------------------------------------------------------

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "D:\Python Leaning\BIKE DETAILS.csv"
bike_data = pd.read_csv(file_path)

# 1. What is the range of selling prices in the dataset?
selling_price_range = bike_data["selling_price"].max() - bike_data["selling_price"].min()

# 2. What is the median selling price for bikes in the dataset
median_selling_price = bike_data["selling_price"].median()

# 3. What is the most common seller type?
most_common_seller_type = bike_data["seller_type"].mode()[0]

# 4. How many bikes have driven more than 50,000 kilometers?
bikes_above_50k_km = (bike_data["km_driven"] > 50000).sum()

# 5. What is the average km_driven value for each ownership type?
avg_km_by_owner = bike_data.groupby("owner")["km_driven"].mean()

# 6. What proportion of bikes are from the year 2015 or older?
bikes_2015_or_older = (bike_data["year"] <= 2015).sum()
proportion_2015_or_older = bikes_2015_or_older / len(bike_data)

# 7. What is the trend of missing values across the dataset?
missing_values = bike_data.isnull().sum()

# 8. What is the highest ex_showroom_price recorded, and for which bike?
highest_ex_showroom_price = bike_data.loc[bike_data["ex_showroom_price"].idxmax(), ["name", "ex_showroom_price"]]

# 9. What is the total number of bikes listed by each seller type?
bikes_by_seller_type = bike_data["seller_type"].value_counts()
(
selling_price_range,
median_selling_price,
    most_common_seller_type,
    bikes_above_50k_km,
    avg_km_by_owner,
    proportion_2015_or_older,
    missing_values,
    highest_ex_showroom_price,
    bikes_by_seller_type
)

# 10. What is the relationship between selling_price and km_driven for first-owner bikes?
# Filter first-owner bikes
first_owner_bikes = bike_data[bike_data["owner"] == "1st owner"]

# Scatter plot for selling price vs km_driven
plt.figure(figsize=(8, 5))
sns.scatterplot(x=first_owner_bikes["km_driven"], y=first_owner_bikes["selling_price"], alpha=0.5)
plt.xlabel("Kilometers Driven")
plt.ylabel("Selling Price (INR)")
plt.title("Relationship between Selling Price and Km Driven for First-Owner Bikes")
plt.show()

#11. identify and remove outliers in the km_driven column using the IQR method?
# Outlier removal using IQR method for km_driven
Q1 = bike_data["km_driven"].quantile(0.25)
Q3 = bike_data["km_driven"].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
filtered_bike_data = bike_data[(bike_data["km_driven"] >= lower_bound) & (bike_data["km_driven"] <= upper_bound)]

# Number of records removed
outliers_removed = len(bike_data) - len(filtered_bike_data)
outliers_removed

#12. Perform a bivariate analysis to visualize the relationship between year and selling_price?
# Bivariate analysis: Year vs Selling Price
plt.figure(figsize=(10, 5))
sns.boxplot(x=filtered_bike_data["year"], y=filtered_bike_data["selling_price"])
plt.xticks(rotation=45)
plt.xlabel("Manufacturing Year")
plt.ylabel("Selling Price (INR)")
plt.title("Bivariate Analysis: Year vs Selling Price")
plt.show()

# 13. What is the average depreciation in selling price based on the bike's age (current year - manufacturing year)?
    # Calculate bike age (current year - manufacturing year)
current_year = 2025
filtered_bike_data["bike_age"] = current_year - filtered_bike_data["year"]
# Calculate average depreciation per year
depreciation_per_year = filtered_bike_data.groupby("bike_age")["selling_price"].mean().diff()
# Display depreciation values
depreciation_per_year

#14. Which bike names are priced significantly above the average price for their manufacturing year?
# Calculate the average selling price for each manufacturing year
avg_price_per_year = filtered_bike_data.groupby("year")["selling_price"].mean()

# Identify bikes priced significantly above the average (greater than 1.5 times the average price for that year)
outlier_bikes = filtered_bike_data[
        filtered_bike_data.apply(lambda x: x["selling_price"] > 1.5 * avg_price_per_year[x["year"]], axis=1)
    ]

# Select relevant columns
outlier_bikes[["name", "year", "selling_price"]]

# 15. Develop a correlation matrix for numeric columns and visualize it using a heatmap.
# Select only numeric columns
numeric_cols = ["selling_price", "year", "km_driven", "ex_showroom_price", "bike_age"]
corr_matrix = filtered_bike_data[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, vmin=-1, vmax=1)
plt.title("Correlation Matrix of Numerical Features")
plt.show()


#EDA-2-------------------------------------------------------

car_sale_file_path = "D:\Python Leaning\Car Sale.csv"
car_data = pd.read_csv(car_sale_file_path)

# 1. What is the average selling price of cars for each dealer?
avg_price_by_dealer = car_data.groupby("Dealer_Name")["Price ($)"].mean().sort_values(ascending=False)
# 2. Which car brand (Company) has the highest variation in prices?
price_variation_by_company = car_data.groupby("Company")["Price ($)"].std().sort_values(ascending=False)

# 3. What is the distribution of car prices for each transmission type?
price_distribution_by_transmission = car_data.groupby("Transmission")["Price ($)"].describe()

# 4. What is the distribution of car prices across different regions?
price_distribution_by_region = car_data.groupby("Dealer_Region")["Price ($)"].describe()

# 5. What is the distribution of cars based on body styles?
car_count_by_body_style = car_data["Body Style"].value_counts()

# Display results
(
    avg_price_by_dealer.head(),
    price_variation_by_company.head(),
    price_distribution_by_transmission,
    price_distribution_by_region,
    car_count_by_body_style
)

# 6. How does the average selling price of cars vary by customer gender and annual income?
avg_price_by_gender = car_data.groupby("Gender")["Price ($)"].mean()

# Segmenting customers into income brackets for analysis
income_bins = [0, 50000, 100000, 200000, 500000, 1000000, car_data["Annual Income"].max()]
income_labels = ["<50K", "50K-100K", "100K-200K", "200K-500K", "500K-1M", "1M+"]
car_data["Income Bracket"] = pd.cut(car_data["Annual Income"], bins=income_bins, labels=income_labels)

# Calculate average car price for each income bracket
avg_price_by_income_bracket = car_data.groupby("Income Bracket")["Price ($)"].mean()

# Display results
avg_price_by_gender, avg_price_by_income_bracket

# 7. What is the distribution of car prices by region, and how does the number of cars sold vary by region?
price_distribution_by_region = car_data.groupby("Dealer_Region")["Price ($)"].describe()
cars_sold_by_region = car_data["Dealer_Region"].value_counts()

# Plot number of cars sold by region
plt.figure(figsize=(8, 5))
sns.barplot(x=cars_sold_by_region.index, y=cars_sold_by_region.values, palette="coolwarm")
plt.xlabel("Region")
plt.ylabel("Number of Cars Sold")
plt.title("Number of Cars Sold by Region")
plt.xticks(rotation=45)
plt.show()

# Display price distribution by region
price_distribution_by_region, cars_sold_by_region

# 8. How does the average car price differ between cars with different engine sizes?
avg_price_by_engine = car_data.groupby("Engine")["Price ($)"].mean().sort_values(ascending=False)

# Plot average price by engine type
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_price_by_engine.index, y=avg_price_by_engine.values, palette="coolwarm")
plt.xlabel("Engine Type")
plt.ylabel("Average Price ($)")
plt.title("Average Car Price by Engine Type")
plt.xticks(rotation=45)
plt.show()

# Display results
avg_price_by_engine

# 9. How do car prices vary based on the customer’s annual income bracket?
# Group by income bracket and calculate average price
avg_price_by_income_bracket = car_data.groupby("Income Bracket")["Price ($)"].mean().sort_values()

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_price_by_income_bracket.index, y=avg_price_by_income_bracket.values, palette="coolwarm")
plt.xlabel("Annual Income Bracket")
plt.ylabel("Average Car Price ($)")
plt.title("Average Car Price by Annual Income Bracket")
plt.show()

# Display results
avg_price_by_income_bracket

#10. What are the top 5 car models with the highest number of sales, and how does their price distribution look?
#Top 5 car models with the highest number of sales
top_5_models = car_data["Model"].value_counts().head(5)

# Get the price distribution for these top 5 models
top_models_data = car_data[car_data["Model"].isin(top_5_models.index)]
price_distribution_top_models = top_models_data.groupby("Model")["Price ($)"].describe()

# Display results
top_5_models, price_distribution_top_models

#11. How does car price vary with engine size across different car colors, and which colors have the highest price variation?
# Group by engine type and color to calculate mean & standard deviation of price
price_variation_by_engine_color = car_data.groupby(["Engine", "Color"])["Price ($)"].agg(["mean", "std"]).unstack()

# Identify colors with highest price variation (sorted by std deviation)
highest_price_variation_colors = car_data.groupby("Color")["Price ($)"].std().sort_values(ascending=False)

# Plot the highest price variation by color
plt.figure(figsize=(8, 5))
sns.barplot(x=highest_price_variation_colors.index, y=highest_price_variation_colors.values, palette="coolwarm")
plt.xlabel("Car Color")
plt.ylabel("Price Variation (Standard Deviation)")
plt.title("Car Colors with Highest Price Variation")
plt.xticks(rotation=45)
plt.show()

# Display results
price_variation_by_engine_color, highest_price_variation_colors


#12. Is there any seasonal trend in car sales based on the date of sale?
# Convert Date column to datetime format
car_data["Date"] = pd.to_datetime(car_data["Date"], errors="coerce")

# Extract month and analyze sales trends
car_data["Month"] = car_data["Date"].dt.month
monthly_sales = car_data["Month"].value_counts().sort_index()

# Plot seasonal trend
plt.figure(figsize=(8, 5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker="o")
plt.xlabel("Month")
plt.ylabel("Number of Cars Sold")
plt.title("Seasonal Trend in Car Sales")
plt.xticks(range(1, 13))
plt.show()

#13. How does the car price distribution change when considering different combinations of body style and transmission type?

plt.figure(figsize=(10, 6))
sns.boxplot(x="Body Style", y="Price ($)", hue="Transmission", data=car_data)
plt.xlabel("Body Style")
plt.ylabel("Car Price ($)")
plt.title("Car Price Distribution by Body Style and Transmission Type")
plt.xticks(rotation=45)
plt.legend(title="Transmission Type")
plt.show()

# Compute summary statistics
price_distribution_by_body_transmission = car_data.groupby(["Body Style", "Transmission"])["Price ($)"].describe()
price_distribution_by_body_transmission

#14. What is the correlation between car price, engine size, and annual income of customers, and how do these features interact?
# Encode Engine Type as numerical values
car_data["Engine_Encoded"] = car_data["Engine"].astype("category").cat.codes

# Select relevant numeric columns for correlation analysis
corr_columns = ["Price ($)", "Annual Income", "Engine_Encoded"]
corr_matrix = car_data[corr_columns].corr()

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, vmin=-1, vmax=1)
plt.title("Correlation Matrix: Car Price, Engine Type, and Annual Income")
plt.show()

# Display correlation values
corr_matrix

#15. How does the average car price vary across different car models and engine types?
# Group by Model and Engine to calculate average price
avg_price_by_model_engine = car_data.groupby(["Model", "Engine"])["Price ($)"].mean().unstack()

# Get the top 10 models with the highest average price
top_10_models = avg_price_by_model_engine.mean(axis=1).sort_values(ascending=False).head(10)

# Plot top 10 car models with average price by engine type
plt.figure(figsize=(10, 5))
top_10_models_data = avg_price_by_model_engine.loc[top_10_models.index]
top_10_models_data.plot(kind="bar", stacked=True, figsize=(10, 5), colormap="coolwarm")
plt.xlabel("Car Model")
plt.ylabel("Average Price ($)")
plt.title("Top 10 Car Models with Highest Average Price by Engine Type")
plt.xticks(rotation=45)
plt.legend(title="Engine Type")
plt.show()

# Display top 10 models with highest average price
top_10_models_data

#EDA-3----------------------------------------------------------

amazon_file_path = r"D:\Python Leaning\amazon.csv"
amazon_data = pd.read_csv(amazon_file_path)

#1. What is the average rating for each product category?
# Ensure all values are strings before replacing symbols
amazon_data["discounted_price"] = amazon_data["discounted_price"].astype(str).str.replace("₹", "").str.replace(",", "").astype(float)
amazon_data["actual_price"] = amazon_data["actual_price"].astype(str).str.replace("₹", "").str.replace(",", "").astype(float)
amazon_data["discount_percentage"] = amazon_data["discount_percentage"].astype(str).str.replace("%", "").astype(float)
amazon_data["rating"] = pd.to_numeric(amazon_data["rating"], errors="coerce")
amazon_data["rating_count"] = amazon_data["rating_count"].astype(str).str.replace(",", "").astype(float)

avg_rating_by_category = amazon_data.groupby("category")["rating"].mean()
print(avg_rating_by_category)

#2. What are the top rating_count products by category ?
top_products_by_category = amazon_data.loc[amazon_data.groupby("category")["rating_count"].idxmax(), ["category", "product_name", "rating_count"]]
print(top_products_by_category)

#3. What is the distribution of discounted prices vs. actual prices?
# Plot the distribution of discounted and actual prices
plt.figure(figsize=(8, 5))
sns.histplot(amazon_data["discounted_price"], color="blue", label="Discounted Price", kde=True)
sns.histplot(amazon_data["actual_price"], color="red", label="Actual Price", kde=True)
plt.xlabel("Price (₹)")
plt.ylabel("Frequency")
plt.legend()
plt.title("Distribution of Discounted Prices vs. Actual Prices")
plt.show()
#4. How does the average discount percentage vary across categories?
avg_discount_by_category = amazon_data.groupby("category")["discount_percentage"].mean().sort_values(ascending=False)

# Plot the results
plt.figure(figsize=(10, 5))
sns.barplot(x=avg_discount_by_category.index, y=avg_discount_by_category.values, palette="coolwarm")
plt.xlabel("Product Category")
plt.ylabel("Average Discount (%)")
plt.xticks(rotation=45, ha="right")
plt.title("Average Discount Percentage by Category")
plt.show()

# Display the computed values
print(avg_discount_by_category)

#5. What are the most popular product names
# Find the most popular products based on rating count
popular_products = amazon_data.sort_values(by="rating_count", ascending=False)[["product_name", "rating_count"]].head(10)
print(popular_products)

#6. What are the most popular product keywords
from collections import Counter
import itertools
# Tokenize and count words in product names
all_words = list(itertools.chain(*amazon_data["product_name"].str.split()))
word_counts = Counter(all_words)
# Get the 10 most common words
most_common_words = word_counts.most_common(10)
print(most_common_words)

#7. What are the most popular product reviews
# Find the most popular reviews based on rating count
popular_reviews = amazon_data.sort_values(by="rating_count", ascending=False)[["review_title", "review_content", "rating_count"]].head(10)
print(popular_reviews)

#8. What is the correlation between discounted_price and rating
# Compute correlation
correlation = amazon_data[["discounted_price", "rating"]].corr()

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, vmin=-1, vmax=1)
plt.title("Correlation Between Discounted Price and Rating")
plt.show()
print(correlation)

#9. What are the Top 5 categories based on the highest ratings
# Compute average rating per category
top_categories_by_rating = amazon_data.groupby("category")["rating"].mean().sort_values(ascending=False).head(5)
print(top_categories_by_rating)

#10. Identify any potential areas for improvement or optimization based on the data analysis.

'''1. Optimize Pricing & Discount Strategies
    🔹 Insight: If there is a strong correlation between discounted price & rating, customers may favor discounted products.

🔹 Optimization:
    Dynamic pricing strategy: Offer discounts on low-rated but high-potential products.
    Bundle pricing: Suggest related products together at a discount to increase sales.
    Personalized offers: Use customer purchase history to offer targeted discounts.
2. Improve Product Descriptions Using Keyword Insights
    🔹 Insight: Common words in popular product names show trending customer preferences.

🔹 Optimization:
    Enhance product titles & descriptions with top-performing keywords.
    SEO-friendly product listings to improve search visibility on Amazon.
    Better product categorization for easy discoverability.
3. Leverage High-Performing Categories for Marketing
    🔹 Insight: Some categories consistently receive higher ratings.

🔹 Optimization:
    Invest in high-rated categories (e.g., electronics, accessories, etc.).
    Run targeted ads focusing on these high-rated categories.
    Expand inventory for in-demand, well-rated categories.
4. Identify & Improve Low-Rated Products
    🔹 Insight: Some products have low ratings despite high sales.
🔹 Optimization:
    Analyze negative reviews to understand pain points.
    Improve product quality based on common complaints.
    Update product images & descriptions to set better expectations.
    Test alternative pricing to see if lower prices increase satisfaction.
5. Promote High-Rated But Low-Reviewed Products
    🔹 Insight: Some products have high ratings but very few reviews.
🔹 Optimization:
    Encourage verified buyers to leave reviews by offering small incentives.
    Feature these products in promotions to gain more visibility.
    Leverage influencer marketing to increase credibility & awareness.
6. Boost Engagement Through Better Customer Reviews
    🔹 Insight: Most popular reviews have detailed feedback & images.

🔹 Optimization:
    Encourage user-generated content (e.g., review images/videos).
    Improve customer interaction by responding to reviews.
    Use AI sentiment analysis to track & address negative feedback faster.
7. Improve Inventory Management Based on Demand
    🔹 Insight: Some products may be frequently out of stock due to high demand.
🔹 Optimization:
    Track top-rated products and ensure consistent stock availability.
    Analyze seasonal trends to manage inventory better.
    Use predictive analytics to avoid stock shortages.
'''

#EDA-4-----------------------------------------------------------

spotify_file_path = "D:\Python Leaning\spotify.csv"
spotify_data = pd.read_csv(spotify_file_path)

#Q1. Read the dataframe, check null value if present then do the needful, check duplicate row , if present then do the needful
# 1. Check for null values
display(spotify_data.isnull().sum())
# Handling missing values
spotify_data.dropna(inplace=True)
#
#2. Check for Duplicate Rows and Remove Them
# Check for duplicates
print("Duplicate Rows: ", spotify_data.duplicated().sum())
# Remove duplicates
spotify_data = spotify_data.drop_duplicates()
# Verify
print("Duplicate Rows After Removal: ", spotify_data.duplicated().sum())


#Q2. What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram Plot popularity distribution
#Interpretion- 
'''
1. This histogram shows how popularity is distributed among the tracks.
2. The KDE (Kernel Density Estimation) curve provides a smooth estimate of the distribution.
3. If the histogram is left-skewed, most tracks have high popularity.
4. If it's right-skewed, most tracks have low popularity.'''

plt.figure(figsize=(10, 5))
sns.histplot(spotify_data['Popularity'], bins=30, kde=True, color="blue")
plt.title("Distribution of Track Popularity")
plt.xlabel("Popularity Score")
plt.ylabel("Count of Tracks")
plt.show()

#Q3. Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot
'''
Interpretation:
1. If there's a clear upward or downward trend, it suggests that track duration affects popularity.
2. If the points are randomly scattered, it means there's little or no correlation.
3. If shorter or longer tracks tend to be more popular, we can further analyze trends using correlation coefficients.
'''

# Scatter plot for Popularity vs. Duration
plt.figure(figsize=(10, 5))
sns.scatterplot(x=spotify_data['Duration (ms)'], y=spotify_data['Popularity'], alpha=0.5, color='red')

# Labels and title
plt.title("Relationship Between Track Duration and Popularity")
plt.xlabel("Duration (ms)")
plt.ylabel("Popularity Score")
plt.show()

#Q4. Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot

# Count the number of tracks per artist
artist_counts = spotify_data['Artist'].value_counts()

# Display the artist with the highest number of tracks
top_artist = artist_counts.idxmax()
top_artist_count = artist_counts.max()
print(f"The artist with the highest number of tracks is {top_artist} with {top_artist_count} tracks.")

# Plot the top 10 artists with most tracks
plt.figure(figsize=(12, 6))
sns.countplot(y=spotify_data['Artist'], order=spotify_data['Artist'].value_counts().index[:10], palette="viridis")

# Labels and title
plt.title("Top 10 Artists with Most Tracks")
plt.xlabel("Number of Tracks")
plt.ylabel("Artist")
plt.show()

#Q5. What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each
# Get the 5 least popular tracks
least_popular_tracks = pd.DataFrame(spotify_data.nsmallest(5, 'Popularity')[['Artist', 'Track Name', 'Popularity']])

# Display the result
print("Top 5 Least Popular Tracks:")
print(least_popular_tracks)

#Q6. Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist

# Find the top 5 most popular artists (those with the most tracks)
top_5_artists = spotify_data['Artist'].value_counts().nlargest(5).index

# Calculate the average popularity for these top 5 artists
average_popularity = spotify_data[spotify_data['Artist'].isin(top_5_artists)].groupby('Artist')['Popularity'].mean()

# Display the results
print("Top 5 Most Popular Artists and Their Average Popularity:\n", average_popularity)

# Identify the artist with the highest average popularity
highest_avg_pop_artist = average_popularity.idxmax()
highest_avg_pop_score = average_popularity.max()
print(f"\nThe artist with the highest average popularity is {highest_avg_pop_artist} with an average popularity of {highest_avg_pop_score:.2f}.")

#Q7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.
# Find the top 5 most popular artists (those with the most tracks)
top_5_artists = spotify_data['Artist'].value_counts().nlargest(5).index

# Find the most popular track for each of these top 5 artists
print("Most Popular Tracks for the Top 5 Artists:\n")
for artist in top_5_artists:
    most_popular_track = spotify_data[spotify_data['Artist'] == artist].nlargest(1, 'Popularity')[['Track Name', 'Popularity']]
    print(f"Most popular track by {artist}: {most_popular_track.values[0][0]} (Popularity: {most_popular_track.values[0][1]})")

#Q8. Visualize relationships between multiple numerical variables simultaneously using a pair plot.
'''
To visualize relationships between multiple numerical variables in the dataset, we can use a pair plot. This will help us identify correlations
and patterns between numerical features like Popularity and Duration (ms).
'''
# Selecting numerical columns for visualization
numerical_columns = ['Popularity', 'Duration (ms)']

# Pair plot to visualize relationships
sns.pairplot(spotify_data[numerical_columns])
plt.show()

'''
Explanation:
1. This pair plot will create scatter plots for every numerical variable against the others.
2. If Popularity and Duration (ms) are correlated, you will see a trend in their scatter plot.
3. If they are randomly distributed, there’s no significant relationship.

'''
#Q9. Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot
'''
To determine if track duration varies significantly across different artists, we will use box plots and violin plots. These visualizations 
help show the distribution, spread, and outliers for track durations by artist.
'''
# Select top 10 artists with the most tracks for better visualization
top_10_artists = spotify_data['Artist'].value_counts().nlargest(10).index
filtered_data = spotify_data[spotify_data['Artist'].isin(top_10_artists)]

# Box Plot for Track Duration by Artist
plt.figure(figsize=(12, 6))
sns.boxplot(x='Artist', y='Duration (ms)', data=filtered_data, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Track Duration Variation Across Top 10 Artists")
plt.xlabel("Artist")
plt.ylabel("Duration (ms)")
plt.show()

# Violin Plot for Track Duration by Artist
plt.figure(figsize=(12, 6))
sns.violinplot(x='Artist', y='Duration (ms)', data=filtered_data, palette="muted")
plt.xticks(rotation=45)
plt.title("Track Duration Distribution Across Top 10 Artists")
plt.xlabel("Artist")
plt.ylabel("Duration (ms)")
plt.show()

"""
Interpretation:
Box Plot:
    1. Wider boxes mean higher variation in duration.
    2. Outliers (dots) represent unusually short or long tracks.
    3. Artists with tight boxes have consistent track durations.
Violin Plot:
    1. Shows distribution shape for each artist.
    2. Wider parts indicate more tracks with that duration.
    3. Symmetric or skewed distributions reveal trends in track lengths.
"""

#Q10. How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.
# Select the top 10 artists with the most tracks for better visualization
top_10_artists = spotify_data['Artist'].value_counts().nlargest(10).index
filtered_data = spotify_data[spotify_data['Artist'].isin(top_10_artists)]

# Swarm Plot for Track Popularity by Artist
plt.figure(figsize=(12, 6))
sns.swarmplot(x='Artist', y='Popularity', data=filtered_data, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Popularity Distribution Across Top 10 Artists (Swarm Plot)")
plt.xlabel("Artist")
plt.ylabel("Popularity Score")
plt.show()

# Violin Plot for Track Popularity by Artist
plt.figure(figsize=(12, 6))
sns.violinplot(x='Artist', y='Popularity', data=filtered_data, palette="muted")
plt.xticks(rotation=45)
plt.title("Popularity Distribution Across Top 10 Artists (Violin Plot)")
plt.xlabel("Artist")
plt.ylabel("Popularity Score")
plt.show()

'''
Interpretation:
Swarm Plot:
    1. Each dot represents a track.
    2. Clustered dots indicate many tracks with similar popularity.
    3. Spread-out dots show high variability in an artist’s track popularity.
Violin Plot:
    1. Shows density of popularity scores.
    2. Wider areas indicate more tracks at a certain popularity level.
    3. Symmetric or skewed shapes reveal trends in track success.

'''
```
